# SQL Where Exercises

```python
import pandas as pd
import sqlite3
import pandasql

conn = sqlite3.connect('flights.db')
cur = conn.cursor()
```

## Exercise 1:
Select all records where the city column has the value "Berlin".

```python
pd.read_sql('''
SELECT *
FROM airports
WHERE city = 'Berlin'
''', conn)
```

## Exercise 2:
Use the NOT keyword to select all records where city is NOT "Berlin".

```python
pd.read_sql('''
SELECT *
FROM airports
WHERE NOT city = 'Berlin'
''', conn)
```

## Exercise 3:
Select all records where the altitude column has the value 0.

```python
pd.read_sql('''
SELECT *
FROM airports
WHERE altitude = 0
''', conn)
```

## Exercise 4:
Select all records where the country column has the value 'Canada' and the altitude column has the value 0.

```python
pd.read_sql('''
SELECT *
FROM airports
WHERE country = 'Canada'
AND altitude = 0
''', conn)
```

## Exercise 5:
Select all records where the city column has the value 'Berlin' or 'London'.

```python
pd.read_sql('''
SELECT *
FROM airports
WHERE city = 'London'
OR city = 'Berlin'
''', conn)
```
