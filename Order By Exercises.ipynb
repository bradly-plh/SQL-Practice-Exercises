# SQL Order By Exercises

```python
import pandas as pd
import sqlite3
import pandasql

conn = sqlite3.connect('flights.db')
cur = conn.cursor()
```

## Exercise 1:
Select al records from the airports table, sort the results alphabetically by column city.

```python
pd.read_sql('''
SELECT *
FROM airports
ORDER BY city
''', conn)
```

## Exercise 2:
Select all records from the airports table, sortthe result reversed alphabetically by the column city.

```python
pd.read_sql('''
SELECT *
FROM airports
ORDER BY city DESC
''', conn)
```

## Exercise 3:
Select all records from te airports table, sort the result alphabetically, first by the column country, then by the column city.

```python
pd.read_sql('''
SELECT *
FROM airports
ORDER BY country, city
''', conn)
```
